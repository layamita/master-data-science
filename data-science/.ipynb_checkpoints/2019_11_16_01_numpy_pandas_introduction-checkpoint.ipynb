# Ejercicio

Y = AX + B donde A=2.5 y B = 20

```python
def func(x):
    a = 2.5
    b = 20
    return (a*x)+b
```

```python
arr_1 = np.linspace(-10,10,10)
plt.plot(arr_1,func(arr_1),"r");
plt.xlim([-20,20])
plt.ylim([-10,60])
```

```python
arr_2 = np.arange(-10,10,2)
plt.plot(arr_2,func(arr_2),"b");
```

```python
arr_3 = (np.random.rand(10))
plt.plot(arr_3,func(arr_2),"g");
```

## Representar la funcion sigmoid entre -20 y 20

```python
num_elem = 100
array_4 = np.linspace(-20,20, num_elem)
```

```python
def sig(x):
    return 1 / (1 + np.exp(-x))
```

```python
plt.scatter(array_4,sig(array_4),color="green");
```
