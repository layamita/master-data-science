Las ventas diarias de una empresa siguen una distribuciÃ³n uniforme entre 300 y 600 euros. Suponiendo independientes las ventas de los distintos dÃ­as del aÃ±o, calcular la probabilidad de que el volumen de ventas anual supere los 138.000 euros, si la empresa trabaja 300 dÃ­as al aÃ±o.

```python
tmp = stats.uniform(loc=300,scale=600-300)
```

```python
x = np.linspace(250,650,500)
plt.plot(x,tmp.pdf(x));
```

```python
suma_unif = stats.norm(300*tmp.mean(),np.sqrt(300)*tmp.std())
```

```python
1-suma_unif.cdf(138000)
```
