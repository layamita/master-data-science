La duraci贸n en minutos de un determinado viaje es una variable aleatoria con distribuci贸n normal de media desconocida y desviacion tipica desconocida. En una muestra aleatoria de diez realizaciones del viaje en cuesti贸n se obtuvieron los siguientes tiempos

1=10,12=6,53=5,54=7,95=8,26=6,5 7=7,08=8,19=6,9 10=7,7.

 

Se pide:Estimar por el m茅todo de los momentos la duraci贸n media del viaje.
Calcular la probabilidad de que, en valor absoluto, la diferencia entre la estimaci贸n media estimada y la real sea menor que un minuto.

```python
%pylab inline
```

```python
from scipy import stats
```

```python
data = np.array([10.1,6.5,5.5,7.9,8.2,6.5,10.1,6.5,5.5,7.9,8.2,6.5,7.0,8.1,6.9,7.7])
```

```python
media = np.mean(data)
```

```python
des = np.std(data)
```

```python
est_norm = stats.norm(media, des)
```

```python
est_norm.mean()
```

```python
plt.hist(data,normed="true")
x = np.linspace(4,11,100)
plt.plot(x,est_norm.pdf(x));
```

#Calcular la probabilidad de que, en valor absoluto, la diferencia entre la estimaci贸n media estimada y la real sea menor que un minuto.

```python
mi_norm = stats.t(9)
```

```python
t1 = ((est_norm.mean()-1)-est_norm.mean())/(des/np.sqrt(10))
```

```python
t2 = ((est_norm.mean()+1)-est_norm.mean())/(des/np.sqrt(10))
```

```python
x=np.linspace(-5,5,500)
plt.plot(x,mi_norm.pdf(x))

x1=np.linspace(-5,t1,500)
plt.fill_between(x1,mi_norm.pdf(x1))

x2=np.linspace(t2,5,500)
plt.fill_between(x2,mi_norm.pdf(x2));
```

```python
1-(mi_norm.cdf(t2)-mi_norm.cdf(t1))
```

cual es el intervalo de confianza del 95% para la media muestral

```python
t2 = mi_norm.ppf(0.975)
t2*(des/np.sqrt(10))+ est_norm.mean()
```
