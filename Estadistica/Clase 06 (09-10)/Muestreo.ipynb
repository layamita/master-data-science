```python
x=np.linspace(0,6,500)
plt.plot(x,mi_norm.pdf(x))

x1=np.linspace(0,2.9,500)
plt.fill_between(x1,mi_norm.pdf(x1))

x2=np.linspace(3.1,6,500)
plt.fill_between(x2,mi_norm.pdf(x2))
```
