```python
# https://www.statsmodels.org/stable/examples/notebooks/generated/kernel_density.html
```

```python
from scipy import stats
```

```python
sample = np.concatenate([stats.norm(2,1).rvs(100),stats.norm(4,.5).rvs(100)])
```

```python
plt.hist(sample,bins=40);
```

```python
# KDE
import statsmodels.api as sm
```

```python
kde = sm.nonparametric.KDEUnivariate(sample)
kde.fit()
```

```python
plt.plot(kde.support, kde.density, lw=3, label='KDE from samples', zorder=10)

plt.scatter(sample,0.01*np.random.random(len(sample)))
```

```python
# True
x = np.linspace(-1.5,6.5,200)
plt.plot(x,(stats.norm(2,1).pdf(x)+stats.norm(4,.5).pdf(x))/2)

# Estimated
plt.plot(kde.support, kde.density,color="red")

# Points
plt.scatter(sample,0.01*np.random.random(len(sample)))
```

```python
plt.plot(kde.support,kde.cdf)
```
